```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Input, Dense, LeakyReLU, Embedding, Flatten, Dropout, Concatenate, Reshape
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

# Load MNIST data
(x_train, y_train), (_, _) = mnist.load_data()
x_train = x_train.astype(np.float32) / 127.5 - 1.0   # Normalize to [-1,1]
x_train = x_train.reshape(-1, 28*28)
num_classes = 10
latent_dim = 100

# Generator
def build_generator():
    label = Input(shape=(1,), dtype='int32')
    label_embedding = Embedding(num_classes, latent_dim, input_length=1)(label)
    label_embedding = Flatten()(label_embedding)

    noise = Input(shape=(latent_dim,))
    model_input = Concatenate()([noise, label_embedding])

    x = Dense(256)(model_input)
    x = LeakyReLU(0.2)(x)
    x = Dense(512)(x)
    x = LeakyReLU(0.2)(x)
    x = Dense(1024)(x)
    x = LeakyReLU(0.2)(x)
    x = Dense(784, activation='tanh')(x)
    img = Reshape((784,))(x)
    model = Model([noise, label], img, name="Generator")
    return model

# Discriminator
def build_discriminator():
    img = Input(shape=(784,))
    label = Input(shape=(1,), dtype='int32')
    label_embedding = Embedding(num_classes, 784, input_length=1)(label)
    label_embedding = Flatten()(label_embedding)
    model_input = Concatenate()([img, label_embedding])

    x = Dense(512)(model_input)
    x = LeakyReLU(0.2)(x)
    x = Dropout(0.4)(x)
    x = Dense(256)(x)
    x = LeakyReLU(0.2)(x)
    x = Dropout(0.4)(x)
    validity = Dense(1, activation='sigmoid')(x)
    model = Model([img, label], validity, name="Discriminator")
    return model

# Optimizer
optimizer = Adam(0.0002, 0.5)
discriminator = build_discriminator()
discriminator.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
generator = build_generator()

# Print architectures
print("Generator architecture:")
generator.summary()
print("\nDiscriminator architecture:")
discriminator.summary()

# Combined model (generator + discriminator for generator training)
noise = Input(shape=(latent_dim,))
label = Input(shape=(1,), dtype='int32')
img = generator([noise, label])
discriminator.trainable = False
valid = discriminator([img, label])
combined = Model([noise, label], valid)
combined.compile(loss='binary_crossentropy', optimizer=optimizer)

# Training
epochs = 10000
batch_size = 128
sample_interval = 2000

d_losses, g_losses = [], []

for epoch in range(1, epochs + 1):
    # Train Discriminator
    idx = np.random.randint(0, x_train.shape[0], batch_size)
    imgs, labels = x_train[idx], y_train[idx].reshape(-1, 1)

    noise = np.random.normal(0, 1, (batch_size, latent_dim))
    gen_labels = np.random.randint(0, num_classes, (batch_size, 1))
    gen_imgs = generator.predict([noise, gen_labels], verbose=0)

    real = np.ones((batch_size, 1))
    fake = np.zeros((batch_size, 1))

    d_loss_real = discriminator.train_on_batch([imgs, labels], real)
    d_loss_fake = discriminator.train_on_batch([gen_imgs, gen_labels], fake)
    d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

    # Train Generator
    noise = np.random.normal(0, 1, (batch_size, latent_dim))
    trick_labels = np.random.randint(0, num_classes, (batch_size, 1))
    g_loss = combined.train_on_batch([noise, trick_labels], real)
    d_losses.append(d_loss[0])
    g_losses.append(g_loss)

    # Print and visualize
    if epoch % sample_interval == 0 or epoch == 1:
        print(f"{epoch} [D loss: {d_loss[0]:.4f} | acc.: {100*d_loss[1]:.2f}%] [G loss: {g_loss:.4f}]")
        # Visualization: grid of 10 digits (one per class)
        noise = np.random.normal(0, 1, (10, latent_dim))
        sample_labels = np.arange(0, 10).reshape(-1, 1)
        gen_imgs = generator.predict([noise, sample_labels], verbose=0)
        gen_imgs = 0.5 * gen_imgs + 0.5
        fig, axs = plt.subplots(1, 10, figsize=(14, 2))
        for i in range(10):
            axs[i].imshow(gen_imgs[i].reshape(28, 28), cmap='gray')
            axs[i].axis('off')
            axs[i].set_title(str(i))
        plt.suptitle(f'Generated Digits at Epoch {epoch}')
        plt.show()

# Plot loss curves
plt.figure(figsize=(10, 5))
plt.plot(d_losses, label="Discriminator loss")
plt.plot(g_losses, label="Generator loss")
plt.title("CGAN Losses during Training")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
